```python
#creating Animals class
class Animal:
    def __init__(self, name, speed, sound, hours_of_sleep):
        self.name = name
        self.speed = speed
        self.sound = sound
        self.hours_of_sleep = hours_of_sleep

#defining methods for animal class    
    def speak(self):
        print (f"{self.name} says {self.sound}")
    
    def running(self):
        print(f"{self.name} is running {self.speed} km per h.")
        
    def sleep(self):
        print(f"{self.name} sleeps for {self.hours_of_sleep} hours a day.")
        
#creating Dogs class 
class Dog(Animal):
    def __init__(self, name, speed, sound, hours_of_sleep, hours_of_walking, food:float):
        super().__init__(name, speed, sound, hours_of_sleep)
        self.hours_of_walking = hours_of_walking
        self.food = food

#defining methods for dogs   
    def speak(self):
        print(f"{self.name} woofs.")
        
    def active(self):
        print(f"{self.name} is active for {self.hours_of_walking} hours a day.")

    def eating(self):
        if self.food == 1:
            print(f"{self.name} wants to eat.")
        else: 
            print(f"{self.name} doesn't want to eat right now.")
        
#creating Cats class
class Cat(Animal):
    def __init__(self, name, speed, sound, hours_of_sleep, eye_color):
        super().__init__(name, speed, sound, hours_of_sleep)
        self.eye_color = eye_color

#defining methods for cats    
    def speak(self):
        print(f"{self.name} meows.")
        
    def playing(self):
        print(f"{self.name} is playing.")
        

class Horse(Animal):
    def __init__(self, name, speed, sound, hours_of_sleep, tail_color):
        super().__init__(name, speed, sound, hours_of_sleep)
        self.tail_color = tail_color
        
    def speak(self):
        print(f"{self.name} neighs.")
        
    def wagging_its_tail(self):
        print(f"{self.name} is wagging its {self.tail_color} tail.")
```

```python
#defining dogs
dog_1 = Dog("Cooper", 20, "woof", 13, 3, 1)
dog_2 = Dog("Felix", 25, "woof", 14, 2, 0)

dog_1.speak()  # Cooper is making a sound.
dog_1.active()  #For how many hours a day Cooper is active.
dog_1.sleep() #How many hours a day Cooper sleeps
dog_1.eating() #Checking if Cooper wants to eat
print(" ")

dog_2.speak() #Felix is making a sound.
dog_2.active() #For how many hours a day Felix is active.
dog_2.sleep() #How many hours a day Felix sleeps.
dog_2.eating() #Checking if Felix wants to eat
print(" ")

#defining cats 
cat_1 = Cat("Oscar", 13, "meow", 16, "green")
cat_2 = Cat("Pixel", 17, "meow", 15, "gray")

cat_1.speak()    #Oscar is making a sound.
cat_1.playing()  #Oscar is playing.
cat_1.running() #Checking what speed Oscar has
print(" ")

cat_2.speak()    #Pixel is making a sound.
cat_2.playing()  #Pixel is playing.
cat_2.running() #Checking what speed Pixel has
print(" ")

#defining horses 
horse_1 = Horse("Spirit", 60, "neigh", 5, "brown")
horse_2 = Horse("Billy", 57, "neigh", 4, "white")

horse_1.speak()  #Spirit neighs.
horse_1.wagging_its_tail() #Spirit is wagging his tail.
horse_1.running()
print(" ")

horse_2.speak()  #Billy neighs.
horse_2.wagging_its_tail() #Billy is wagging his tail.
horse_2.running()
print(" ")
```
